```python
import copy

def add_par(par):
    newpar = copy.copy(par)
    for x in par:
        newstr = "("+x+")"
        newpar.add(newstr)
        newstr = "()"+x
        newpar.add(newstr)
        newstr = x+"()"
        newpar.add(newstr)
    return newpar
        
```

```python
import itertools

def solution(n):
    answer = 0
    canlist = set(map(''.join, itertools.product("()", repeat=2*n)))
    allowlist = set()
    for x in canlist:
        m=0
        for c in x:
            if c=='(':
                m+=1
            elif c==')':
                m-=1
            if m<0:
                break
        if m==0:
            allowlist.add(x)
#     print(len(allowlist), allowlist)    
    answer =len(allowlist)
    return answer
```

```python
for i in range(1,14):
    print(solution(i))
```

```python

# value= (, ) 증감.
# length=
# n = parenthesis count (input)
brcnt=0
callcnt=0

def make_br(st, o, c, n):
    global brcnt, callcnt
    callcnt+=1
    prev = callcnt
    l = len(st)
    if l==n*2:
        print(st)
        brcnt+=1
        return
    if o<n:
        print(callcnt, prev,'st=', st, o, c, n, 'add (')
        newst = '('+st
        make_br(newst, o+1, c, n)
    if o>c:
        print(callcnt, prev, 'st=', st, o, c, n, 'add )')
        newst = st+')'
        make_br(newst, o, c+1, n)
    
```

```python
make_br('', 0, 0,  3)
print(brcnt)
```
