```python
stack1=[]
stack2=[]
stack3=[]

def gettop(stackindex):
    if stackindex==1:
        if len(stack1)==0:
            return -1
        return stack1[len(stack1)-1]
    if stackindex==2:
        if len(stack2)==0:
            return -1
        return stack2[len(stack2)-1]
    if stackindex==3:
        if len(stack3)==0:
            return -1
        return stack3[len(stack3)-1]
    return None

def checkvalid(fromidx, toidx):
    v1 = gettop(fromidx)
    v2 = gettop(toidx)
    if v1==-1:
        return False
    if v2==-1:
        return True
    if v1<v2:
        return True
    return False


def removeset(setvar, p):
    delobj = []
    if p==1:
        delobj=[(1,2),(1,3)]
    elif p==2:
        delobj=[(2,1),(2,3)]
    else:
        delobj=[(3,1),(3,2)]
    for o in delobj:
        if o in setvar:
            setvar.remove(o)


def hanoi(move):
    global stack1, stack2, stack3
    
    while True:
        movecase = copy.copy(moveall)
        f, t = 0, 0
        notf1, notf2, notf3 = False, False, False
        print('stack')
        print(stack1, stack2, stack3, move)

        # solve 여부 체크.
        if len(stack1)==0 and len(stack2)==0:
            print("Solve", move)
            break
            
        # 시작 단계. 스택 1, 2 둘 중 하나가 빈 다면? 시작스택 설정.
        if len(stack1)==0:
            notf1=True
            if len(stack2)%2==0: # 짝수개. other
                f=2
                t=1
            else:
                f=2
                t=3
        elif len(stack2)==0:
            notf2=True
            if len(stack1)%2==0: # 짝수개. other
                f=1
                t=2
            else:
                f=1
                t=3
        if len(stack3)==0:
            notf3=True

        if notf1:
            removeset(movecase, 1)
        if notf2:
            removeset(movecase, 2)
        if notf3:
            removeset(movecase, 3)
            
        # 둘 다 갈곳이 없다면?
        if f==1:
            if not checkvalid(1,2) and not checkvalid(1,3):
                notf1=True
        if f==2:
            if not checkvalid(2,1) and not checkvalid(2,3):
                notf2=True
        if f==3:
            if not checkvalid(3,1) and not checkvalid(3,2):
                notf3=True
        
        if notf1:
            removeset(movecase, 1)
        if notf2:
            removeset(movecase, 2)
        if notf3:
            removeset(movecase, 3)

        if len(move)>0:
            # 마지막 움직임.
            lastmove = move[len(move)-1]
            removeset(movecase, lastmove)
            removeset(movecase, (lastmove[1], lastmove[0]))
        if len(move)>0 and f==0:
            lastmove[1] # 마지막 움직임에서 도착지 스택은. 출발지가 될 수 없다.
            # 즉 이 값은 f가 될 수 없다.
            if lastmove[1]==1:
                notf1 = True
            elif lastmove[1]==2:
                notf2 = True
            elif lastmove[1]==3:
                notf3 = True
        
        if notf1:
            removeset(movecase, 1)
        if notf2:
            removeset(movecase, 2)
        if notf3:
            removeset(movecase, 3)
            
        
        print('cases=', movecase)
        if notf1 and notf2:
            f=3
        if notf1 and notf3:
            f=2
        if notf2 and notf3:
            f=1
            
        print('debug2 f=', f, 't=', t)

        if f==1 :
            val = stack1.pop()
        elif f==2:
            val = stack2.pop()
        elif f==3:
            val = stack3.pop()
            
        if t==0:
            print('debug3 t not found.')
            if f==1:
                if val == stack2[len(stack2)-1]-1:
                    t=2
                if val == stack3[len(stack3)-1]-1:
                    t=3
            if f==2:
                if val == stack1[len(stack1)-1]-1:
                    t=1
                if val == stack3[len(stack3)-1]-1:
                    t=3
            if f==3:
                if val == stack2[len(stack2)-1]-1:
                    t=2
                if val == stack1[len(stack1)-1]-1:
                    t=1
                    
            if t==0:
                if f==1:
                    if stack2[len(stack2)-1] > stack3[len(stack3)-1]:
                        t=2
                    else:
                        t=3
                if f==2:
                    if stack1[len(stack1)-1] > stack3[len(stack3)-1]:
                        t=1
                    else:
                        t=3
                if f==3:
                    if stack2[len(stack2)-1] > stack1[len(stack1)-1]:
                        t=2
                    else:
                        t=1
            print('debug4 t=', t)
            
        print('debug5 f=', f, 't=', t)
        
        if f==0 or t==0:
            break
        
        move.append((f,t))
        if t==1:
            stack1.append(val)
        if t==2:
            stack2.append(val)
        if t==3:
            stack3.append(val)
        
        
 

            
            
    
```

```python
inf=999999

# 각 스택의 top 원소
def getTop(stack):
    stacktop=[inf,inf,inf]
    for i in range(3):
        if len(stack[i])>0:
            stacktop[i]=stack[i][len(stack[i])-1]
    return stacktop

# 스택별 top에서 부터 연속인 최대값
def getMaxSeq(stack):
    stackmaxseq=[0,0,0]
    stackmaxseqlen=[0,0,0]
    for k,s in enumerate(stack):
        if len(s)>0:
            for i in range(len(s)-1, -1, -1):
                if stackmaxseq[k]+1==s[i]:
                    stackmaxseq[k]=s[i]
                    stackmaxseqlen[k]+=1
                else:
                    break
    return stackmaxseq, stackmaxseqlen
    

# stack is list of list
def isComplete(stack):
    if len(stack[0])==0 and len(stack[1])==0:
        return True
    return False

# prevmove= (0,1)  ;from to.  없으면 None
def nextMove(stack, prevmove):
    fro=-1
    to=-1
    target=2 # 0,1,2 스택번호로 보고. 2번을 타겟으로 한다.
    if prevmove==None:
        prevmove=(-1,-1)
        
#     print('\n stack=', stack, 'prev=', prevmove)
    
    # from 선택
    '''
        나머지 두 개가 empty이면 원소가 있는 스택만 움직일 수 있다. 선정.
        이전 움직임이 (m,n)이면 다음 움직임에서 from은 n일 수 없다. 나머지 두 군데 중 하나가 from.
            나머지 두개의 top이 나의 top보다 작은 경우. 움직일 수 없음.
            이전 움직임의 역과정은 이루어질 수 없다. 즉 (m,n)인데 다음에 (n,m)일 수 없다. (무한루프)
        이전 from을 선택.
        움직일 수 있는지 체크. 이전 from에서 다른 두 곳 중 하나라도 이동가능한가? OK. 선정.  한 곳밖에 못가면 to 선정.
        이전 from을 못움직이면 나머지 하나를 from으로 선정.    
    '''
    stacktop = getTop(stack)
    stackmaxseq, stackmaxseqlen = getMaxSeq(stack)
    availmove = [True,True,True]
    for i in range(3):
        if len(stack[i])==0:
            availmove[i]=False
            
    if prevmove[1]>=0:
        availmove[prevmove[1]]=False
        
    if sum(availmove)==1: # only one
        fro = availmove.index(True)
        
    # not yet
    if fro==-1:
        fro=prevmove[0]
        # can move
        order=sorted(range(len(stacktop)), key=lambda x : stacktop[x]) 
        # [0,1,2] 오름차순으로 정렬.stacktop[0]<stacktop[1]<stacktop[2]
        whichme = order.index( fro )
        if whichme==2 : # 이전 시작위치의 탑이 가장 크면 불능.
            availmove[fro]=False
            fro=-1
        elif whichme==1: # 한 곳 밖에 못 감.
            to = order[2]
        elif whichme==0: # to는 두 곳의 선택지.
            pass
        
    if sum(availmove)==1: # only one
        fro = availmove.index(True)
    
    # to 선택
    '''
    # 4. to 선택.(아직 미정인 경우.)
        둘 중 하나만 이동 가능한 경우. (크기비교). to 선정.
        둘 다 가능한 경우???  연속인 경우를 선택.  
            연속이 없는 경우? 탑의 개수와 목적지 방식. 
            목적지는? 현재탑의 bottom과 연속인 스택의 top.     
    '''
    if to==-1:
        # can move
        order=sorted(range(len(stacktop)), key=lambda x : stacktop[x]) 
        # [0,1,2] 오름차순으로 정렬.stacktop[0]<stacktop[1]<stacktop[2]
        whichme = order.index( fro )
        if whichme==2 : # from이 가장 크면 불능.
            print('cannotmove Error!')
        elif whichme==1: # 한 곳 밖에 못 감.
            to = order[2]
        elif whichme==0: # to는 두 곳의 선택지.
            # 연속 탑.
            sm = stackmaxseq[fro]  # from 연속탑의 바닥이 다른곳의 top과 연결.
            if sm==stacktop[0]-1:
                target=0
            elif sm==stacktop[1]-1:
                target=1
            elif sm==stacktop[2]-1:
                target=2
            fromcount = stackmaxseqlen[fro]
#             print('sm=',sm, target, fromcount)
            if fro==target:
                print('fro=target Error!')
            if fromcount%2==1 : # 홀수개면 타겟으로
                to = target
            else: # 짝수면 other
                to = list({0,1,2} - {fro, target})[0]  
            
        
    return (fro,to)
    
```

```python
def solution(n):
    answer=[]
    stack = [[],[],[]]
    stack[0] = list(range(n, 0, -1))
    m=None
    movecnt=0
    for i in range(40000):
        if isComplete(stack):
            print('Solve')
            print(stack)
            break
        m = nextMove(stack, m)
        movecnt+=1
#         print(movecnt, m)
        answer.append([m[0]+1, m[1]+1])
        v = stack[m[0]].pop(len(stack[m[0]])-1)
        stack[m[1]].append(v)
    print('last movecnt=', movecnt)
#     print(answer)
    return answer
    
solution(15)
```
