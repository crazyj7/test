```python
def make_part(n):
    if n==2:
        return 3
    if n==4:
        return 11
    sum=0 
    for i in range(2, n, 2):
        sum+=make_part(i)*make_part(n-i)
    return sum
```

```python
def tile(n):
    sol=[]
    if n%2!=0:
        return 0
    if n==2:
        return 3
    sol.append(3)
    idx = n//2
    s0=1
    s1=1+3
    for i in range(1, idx):
        print('index=', i-1, s0, s1)
        cnt=(sol[i-1]*3+s0*2) % 1000000007
        print('count=', cnt)
        sol.append(cnt)
        s0=s1
        s1=s1+cnt
    return sol[len(sol)-1] % 1000000007

def solution(n):
    return tile(n)
```
