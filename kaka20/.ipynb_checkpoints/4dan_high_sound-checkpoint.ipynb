```python
# set를 사용하여 중복 추가를 제거함
def make_pat(depth):
    if depth==1:
        return ['*++']
    ret1 = set(map(lambda x : '*'+x+'++', list(make_pat(depth-1))))
    ret2 = set(map(lambda x: '*+'+x+'+', list(make_pat(depth-1))))
    ret3 = set(map(lambda x: '*++'+x, list(make_pat(depth-1))))
    ret0 = set(map(lambda x : x+'*++', list(make_pat(depth-1))))
    return ret1.union(ret2.union(ret3.union(ret0)))


def calc_pat(result):
    cal=[]
    for x in result:
        val=1
        for k in x:
            if k=='*':
                val *= 3
            elif k=='+':
                val += 1
        cal.append(val)
    return cal

# 위 두 함수를 병합
def make_calc(depth, value):
    if deth==1:
        return value*3+2
    ret1 = value*3+2
    ret2 = 1*3+1+value+1
    ret3 = 1*3+2+value

def find_n(value):
    mi,mx=0,0
    n=1
    while True:
        b=3**n
        mi = b+2*n
        mx = b
        for i in range(n):
            mx += 2*(3**i)
#         print (mi, mx)
        if mx>value:
            break
        n+=1
    if mi>value:
        return 0 # not make!
    return n


def calc_x(x):
    val=1
    for k in x:
        if k=='*':
            val *= 3
        elif k=='+':
            val += 1
    return val


def make_pat3(level, n):
    star=list(range(level)) # 0,1,2 ; level=3
    answer=0
    '''
    *++**++... 
    '''

    # 빠른 스캔. 앞에서부터 1,2칸씩 slide
    oldstar=star[:]
    endi=0
    print('scan')
    while True:
        pat=['+']*3*level  
        for s in star:
            pat[s]='*'
        v=calc_x(pat)
        print(''.join(pat), v)
        if v>=n:
            break
        oldstar=star[:]
        bfound=False
        for i in range(level):
            if pat[:star[i]].count('+')<2*i:
                bfound=True
                endi=i
                for j in range(i, level):
                    star[j]+=1
                break
        if bfound==False:
            break
    star=oldstar[:]
    
    print('search')
    while True:
        pat=['+']*3*level  
        for s in star:
            pat[s]='*'
        v=calc_x(pat)
        print(''.join(pat), v)
        if n==v:
            answer+=1
        bfound=False
        for i in range(level-1, 0, -1):
            if i==endi-1:  # ?? endi-1 or endi
                break
            # 뒤에서 부터 움직일 수 있는지 검사한다.
            # pat[2]=5. 그전에 +가 몇 개 인가?
            if pat[:star[i]].count('+') < 2*i :
                bfound=True
                # 뒤에 있는 별들을 당겨온다.
                star[i]+=1
                for j in range(i+1,level):
                    star[j]=star[i]+j-i
                break
        if bfound==False:
            break
    return answer

def solution(n):
    level=find_n(n)
    if level==0:
        return 0
    
#     pat = make_pat(level)
#     cal = calc_pat(pat)
#     answer = cal.count(n)
    answer = make_pat3(level, n)
    print(answer)
    return answer
```
