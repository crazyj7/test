## edge list를 graph map 타입으로

```python
# edge=[fisrt, second] or
# edge =[first, second, cost]
def edge2graph(edges):
    g = dict()
    bcost = False
    if len(edges[0])==3:
        bcost=True
    cost=dict()
    for e in edges:
        if e[0] in g:
            g[e[0]].append(e[1])
            if bcost:
                cost[e[0]].append(e[2])
        else:
            g[e[0]]=[e[1]]
            if bcost:
                cost[e[0]]=[e[2]]
        if e[1] in g:
            g[e[1]].append(e[0])
            if bcost:
                cost[e[1]].append(e[2])
        else:
            g[e[1]]=[e[0]]
            if bcost:
                cost[e[1]]=[e[2]]
    return g, cost
```

## 그래프 그리기 함수 생성

```python
from graphviz import Graph
from PIL import Image
from matplotlib import pyplot as plt

%matplotlib inline

# 무방향 그래프. 중복라인 제거.
def show_graph(graph):
    nodes = list(map(str, graph.keys()))
    dot = Graph(comment='test graph', format='png')
    for n in nodes:
        dot.node(n)
    visit=[]
    for k, v in graph.items():
        for el in v:
            if (el,k) in visit: # 중복스킵
                continue
            dot.edge(str(k), str(el))
            visit.append((k,el))
    dot.render('tmpimage.gv')
    img=Image.open('tmpimage.gv.png')
#     print(img.size) # (가로,세로) 
#     plt.imshow(img)  # 크기가 맞지 않으면 해상도가 좀 깨짐.
    # ipython에서 리턴결과 출력하는 것이 해상도에 맞게 잘 나옴.
    return img  

def show_graph_edge(edges):
    dot = Graph(comment='test graph', format='png')
    for e in edges:
        if len(e)>2:
            dot.edge(str(e[0]), str(e[1]), label=str(e[2]))
        else:
            dot.edge(str(e[0]), str(e[1]))
    
    dot.render('tmpimage.gv')
    img=Image.open('tmpimage.gv.png')
    return img  
```

## BFS (breadth first search) 너비 우선 탐색.
 그래프를 넓게 탐색한다.
큐를 사용한다. 방문여부체크

```python
def bfs(graph, start):
    Q=[]  # 큐 
    V=set() # visit한 노드
    Q.append(start) # 시작 노드를 큐에 추가.
    while len(Q)>0:
        # pop
        cur = Q[0]
        del Q[0]
        if cur in V:
            print('Cycle graph!') # 이미 방문한 노드를 다시 방문!
            continue
        V.add(cur)
        print("Visit=", cur)
        # 연결된 노드 추가
        for e in graph[cur]:
            # Visit한 것은 추가하지 않는다.
            if e not in V:
                Q.append(e)
                print('Q=', Q)

                
# 사이클이 되는지 검사
def cycle_bfs(graph):
    Q=[]  # 큐 
    V=set() # visit한 노드
    start = list(graph.keys())[0]
    Q.append(start) # 시작 노드를 큐에 추가.
    while len(Q)>0:
        # pop
        cur = Q[0]
        del Q[0]
        if cur in V:
            return True
        V.add(cur)
        # 연결된 노드 추가
        for e in graph[cur]:
            # Visit한 것은 추가하지 않는다.
            if e not in V:
                Q.append(e)
    return False
```

## DFS (depth first search) 깊이 우선 탐색
그래프를 깊게 탐색한다. 스택을 이용  
미로에서 한 손을 벽에 대고 탈출하는 방식

```python
# start에 값이 있으면 초기화.
def dfs(graph, dfs_stack, dfs_visit, start):
    if start!=None:
        dfs_stack=[start]
        dfs_visit=set()
        start=None
    if len(dfs_stack)==0:
        return
    # pop
    cur = dfs_stack[-1]
    del dfs_stack[-1]
    print('Visit=', cur)
    if cur in dfs_visit:
        print('Cycle graph!') # 이미 방문한 노드를 다시 방문!
    dfs_visit.add(cur)
    # 연결 노드 추가
    for e in graph[cur]:
        if e not in dfs_visit:
            # 미방문 노드만 추가한다.
            dfs_stack.append(e)
    return dfs(graph, dfs_stack, dfs_visit, None)
```

### Prim  

큐에 푸시  
팝 노드    
연결도로 푸시  
사이클을 만드는 노드로 가는 다리 삭제  
오름 차순 정렬  
가장 싼 다리 선택  
다시 반복하여 팝  


```python
# edge ; [from, to, cost]
def prim(edges):
    
    gmap, cost = edge2graph(edges)
    result = [] # result edges
    visit = []  # visit node
    q=[[edges[0][0], edges[0][0], 0]]  # first node
    while len(q)>0:
        # pop
        # 이미 방문한 노드는 삭제.
        while len(q)>0:
            if q[0][1] not in visit:
                break
            del q[0]
        if len(q)==0:
            break
        node = q[0][1]
        result.append(q[0])
        del q[0]
        
        print('cur node=', node)
        visit.append(node)
        print('visit=', visit)
        
        # add link. 큐에 연결(방문은 제외)노드를 추가하고, 비용순으로 정렬. 
        for n, c in zip(gmap[node], cost[node]):
            # cycle check.
            if n in visit:
                continue
            q.append( [node, n, c] )
        q.sort(key=lambda x: x[2])
        print('q=', q)
        print('result=', result)
    del result[0]
    print(result)
    return result
```

## Kruskal

비용으로 정렬하여 낮은 순서대로 추가한다.   
이때 cycle이 되는 것은 제외하면서 추가한다.  
다른 그룹에 연결되면 그룹 머지를 한다.  

```python
# v가 어느 그룹에 있는지 검색
def find_group(gr, v):
    for i,gi in enumerate(gr):
        if v in gi:
            return i
    return -1
    
def kruskal(g):
    g.sort(key=lambda x:x[2])
    visit=set()
    group = []
    result = []
    for a,b,c in g:
        ia = find_group(group, a)
        ib = find_group(group, b)
        if ia==-1 and ib==-1:
            # 신규
            group.append(set({a,b}))
            result.append([a,b,c])
        elif ia!=-1 and ib==-1:
            # a만 그룹에서 발견
            group[ia].add(b)
            result.append([a,b,c])
        elif ia==-1 and ib!=-1:
            # b만 그룹에서 발견
            group[ib].add(a)
            result.append([a,b,c])
        elif ia!=ib:
            # a,b 모두 다른 그룹. 그룹 머지.
            group[ia]=group[ia].union(group[ib])
            del group[ib]
            result.append([a,b,c])
        else:
            # ia,ib가 같은 그룹. cycle이 되지 않도록 스킵.
            pass
        print('group=', group)
    print(result)
            
            
```

## Dijkstra
최단 거리 찾기.  
최소 비용을 누적해 나간다.  

```python
import copy

def dijkstra(connectmap):
    house=len(connectmap)
    inf = 1e10
    cost = dict()
    # 1번 방을 시작.
    for i in connectmap.keys():
        cost[i] = inf
        
    # 시작을 1번으로 함.
    cost[1] = 0
    # 미방문 코스트
    remaincost = copy.copy(cost)
    
    for i in range(house):
        # 방문할 노드. 미방문중 최소 비용.
        mincost = min(remaincost.items(), key=lambda x : x[1])
        print('visit : mincost=', mincost)
        for l in connectmap[ mincost[0] ]:
            # 방문 제외?
            print(mincost, l)
            c = mincost[1]+1  # 누적 코스트. 단순 거리 측정은 cost를 1로 고정.
            if cost[l]>c:
                # update
                print('update cost', l, c)
                cost[l]=c
                remaincost[l]=c
                print('cost=', cost)
        del remaincost[mincost[0]] # 방문노드 삭제
#         break
    
    return cost    
```

```python
import copy

idebug=0

# edge=[fisrt, second] or
def edge2graph1(edges):
    g = dict()
    # 양방향 모두 추가함.
    for e in edges:
        if e[0] in g:
            g[e[0]].append(e[1])
        else:
            g[e[0]]=[e[1]]
        if e[1] in g:
            g[e[1]].append(e[0])
        else:
            g[e[1]]=[e[0]]
    return g


# edge ; [from, to, cost]
def prim(edges):
    edges.sort()
    if idebug>0:
        print(edges)
    gmap = edge2graph1(edges)
        
    if idebug>0:
        print(gmap)
    result = [] # result edges
    visit = set()  # visit node
    q=[[edges[0][0], edges[0][0]]]  # first node. maybe [1,1]
    if idebug>0:
        print(q)
    while len(q)>0:
        # pop
        # 이미 방문한 노드는 삭제.
        while len(q)>0:
            if q[0][1] not in visit:
                break
            del q[0]
        if len(q)==0:
            break
        node = q[0][1]
        result.append(q[0])
        del q[0]
        
        if idebug>0:
            print('cur node=', node)
        visit.add(node)
        if idebug>0:
            print('visit=', visit)
        
        # add link. 큐에 연결(방문은 제외)노드를 추가하고, 비용순으로 정렬. 
        for n in gmap[node]:
            # cycle check.
            if n in visit:
                continue
            q.append( [node, n] )
        if idebug>0:
            print('q=', q)
            print('result=', result)
    del result[0]
    if idebug>0:
        print(result)
    return result


def solution(n, g):
    inf = 1e10
    result = prim(g)
    if idebug>0:
        print(result)
    # MST
    # calc depth
    parent=dict()
    for e in result:
        parent[e[1]] = e[0]
    depth=dict()
    depth[1]=0
    maxdepth = 0
    maxlist = []
    for i in range(2, n+1):
        if i not in parent or parent[i]==inf :
            depth[i]=inf
        else:
            depth[i] = depth[parent[i]]+1
        if depth[i]>maxdepth:
            maxdepth = depth[i]
            maxlist=[i]
        elif depth[i]==maxdepth:
            maxlist.append(i)
    if idebug>0:
        print(depth)
        print(maxlist)
    answer = len(maxlist)
        
    return answer
```
