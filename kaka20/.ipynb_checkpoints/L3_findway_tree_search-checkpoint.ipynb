```python
class node:
    def __init__(self, index, value):
        self.index = index
        self.value = value
        self.left=-1
        self.right=-1
    def print(self):
        print(self.index, self.value, self.left, self.right)
    def print2(self):
        print(self.index+1, self.value, self.left+1, self.right+1)
    
```

```python
# parent_index 지정.
# nodes는 노드 전체 정보. 리스트
# undernodes ; parent보다 낮은 노드들. 리스트
def make_tree(pidx, nodes, undernodes):
    if len(undernodes)==0:
        return
    
    print(pidx, undernodes)
    idx = sorted( range(len(undernodes)), key=lambda x : -undernodes[x].value[1] )
    maxvalue = undernodes[idx[0]].value[1]
    nodes[pidx].print()
    print('pidx=', pidx,'y=',nodes[pidx].value[1], 'undermaxvalue=', maxvalue)
    print('leaf node')
    undernodes[idx[0]].print()
    
    bleft = False
    if undernodes[idx[0]].value[0] < nodes[pidx].value[0]:
        nodes[pidx].left = undernodes[idx[0]].index
        bleft=True
        print('left!')
    else:
        nodes[pidx].right = undernodes[idx[0]].index
        print('right!')

    maxvalue2=-1
    if len(undernodes)>1:
        maxvalue2 = undernodes[idx[1]].value[1]
        print('maxvalue2=', maxvalue2)
        if maxvalue==maxvalue2:
            if bleft:
                nodes[pidx].right = undernodes[idx[1]].index
            else:
                nodes[pidx].left = undernodes[idx[1]].index

    if nodes[pidx].left>=0:
        undernodes2 = [ node for node in undernodes if node.value[0]<nodes[pidx].value[0] and node.value[1]<maxvalue ]
        make_tree(nodes[pidx].left, nodes, undernodes2)
    if nodes[pidx].right>=0:
        undernodes3 = [ node for node in undernodes if node.value[0]>nodes[pidx].value[0] and node.value[1]<maxvalue ]
        make_tree(nodes[pidx].right, nodes, undernodes3)
       
    
```

```python
def tree_search_fore(nodes, root_index, outlist):
    outlist.append(root_index)
    if nodes[root_index].left>=0:
        tree_search_fore(nodes, nodes[root_index].left, outlist)
    if nodes[root_index].right>=0:
        tree_search_fore(nodes, nodes[root_index].right, outlist)
    return outlist

def tree_search_back(nodes, root_index, outlist):
    if nodes[root_index].left>=0:
        tree_search_back(nodes, nodes[root_index].left, outlist)
    if nodes[root_index].right>=0:
        tree_search_back(nodes, nodes[root_index].right, outlist)
    outlist.append(root_index)
    return outlist
```
