```python
# 괄호를 n쌍으로 표현 가능한 경우의 수
# n=1이면 () 1개
# n=2, (()), ()() 2개   =    n1*3 - 1
# n=3, (n=2) , ()n=2,  n=2() -1 ; 2*3-1 = 5개  = n2*3 - 1  = n2*3 - n1
#     ( (())), (()()),  ()(()),()()(), (())(), ()()().  마지막것이 중복. 
#     ( (())), (()()),  ()(()),()()(), (())()
# n=4이면. (n=3), ()n=3, n=3(), -2 ; 5*3-2=13개  = n3*3 -2  = n3*3 - n2
#     (  ( (())), (()()),  ()(()),()()(), (())()  )   
#   () *  ( (())), (()()),  ()(()),()()(), (())() 
#         ( (())), (()()),  ()(()),()()(), (())()  * ()
# 중복 ; ()를 오른쪽에 붙인 것 중에서, 왼쪽에서 ()로 시작하는 것. ()(()), ()()(), ; 2개가 중복. 
# 
```

```python
#   () *  ( (())), (()()),  ()(()),()()(), (())() 
```

```python
sol=[1,2]

for i in range(3,10):
    ans = sol[i-2]*3-sol[i-3]
    sol.append(ans)
print(sol)
    
```

```python
par = set()
```

```python
par.add( "()" )
```

```python
import copy

def add_par(par):
    newpar = copy.copy(par)
    for x in par:
        newstr = "("+x+")"
        newpar.add(newstr)
        newstr = "()"+x
        newpar.add(newstr)
        newstr = x+"()"
        newpar.add(newstr)
    return newpar
        
```

```python
par = add_par(par)
print(len(par), par)
```

```python
par = add_par(par)
print(len(par), par)
```
